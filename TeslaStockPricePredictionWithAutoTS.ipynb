```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
from datetime import date, timedelta

```

```python
d1 = date.today().strftime("%Y-%m-%d")
end_date = d1
```

```python
d2 = date.today() - timedelta(days=1080)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
df = yf.download('TSLA',
                 start= start_date,
                 end= end_date,
                 progress=False)
print(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format= True)
```

```python
import plotly.graph_objects as go
```

```python
figure = go.Figure(data = [go.Candlestick(x = df.index,
                            open = df["Open"],
                            high= df["High"],
                            low = df["Low"],
                            close = df["Close"])])
figure.update_layout(title = "Time Series Anlaysis(Candlestick Chart with buttons and slider)")

figure.update_xaxes(
                    rangeslider_visible = True,
                    rangeselector = dict(
                        buttons = list([
                            dict(count = 1, label = "1s", step = "second", stepmode = "backward"),
                            dict(count = 1, label = "1m", step = "minute", stepmode = "backward"),
                            dict(count = 1, label = "1h", step = "hour", stepmode = "backward"),
                            dict(count = 1, label = "1d", step = "day", stepmode = "backward"),
                            dict(count = 1, label = "1m", step = "month", stepmode = "backward"),
                            dict(count = 6, label = "6m", step = "month", stepmode = "backward"),
                            dict(count = 1, label = "YTD", step = "year", stepmode = "todate"),
                            dict(count = 1, label = "1y", step = "year", stepmode = "backward"),
                            dict(step = "all")
                    ])
)
)

figure.show()
```

```python
from autots import AutoTS
```

```python
model = AutoTS(forecast_length= 10,
               frequency= 'infer',
               ensemble= 'simple')

model = model.fit(df, date_col = 'Date', value_col = 'Close', id_col = None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
